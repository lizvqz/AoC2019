```python
# maybe save it in a dictionary and each key will have a tuple with x and y value as coordinates {0: (0,0), 1: (8,0)
# , 2: (8,5)}
# save all the 'steps' 0,0; 1,0; 2,0;...8,0; 8,1; 8,2; ...8,5;
# 1,0; 2,0; ...7,0; 7,1; 
# get all the intersections
# get the distance from central point for each intersection.
filename = "/Users/liz/Documents/Advent2019/day3"
distance = []
distances = []

with open(filename) as file:
    for lines in file:
        for rows in lines.strip().split(','):
            distance.append(rows)
        distances.append(distance[:])
        distance.clear()
        
print(distances[0][0][0])

# (0,0), (1,0), (2,0)
            
            
```

```python
combs = []
x_last = 0
y_last = 0

for step in distances[0]:
    if step[0][0] == 'R':
        print(step)
        x_dist = int(step[1:])
        print(x_dist)
        if len(combs) != 0:
            x_last = combs[len(combs)-1][0]
            y_last = combs[len(combs)-1][1]

        for x in range (x_last+1, x_last+x_dist+1): # need to start from last position of x plus one 
            combs.append((x , y_last))
        print(len(combs))
        
    elif step[0][0] == 'L':
        print(step)
        x_dist = int(step[1:])
        print(x_dist)
        if len(combs) != 0:
            x_last = combs[len(combs)-1][0]
            y_last = combs[len(combs)-1][1]

        for x in range (1, x_dist+1): # need to start from last position of x plus one 
            combs.append((x_last-x , y_last))
        print(len(combs))
        
    elif step[0][0] == 'U':
        print(step)
        y_dist = int(step[1:])
        print(y_dist)
        if len(combs) != 0:
            x_last = combs[len(combs)-1][0]
            y_last = combs[len(combs)-1][1]
        for y in range(1, y_dist+1):
            combs.append((x_last, y_last+y))
        print(len(combs))

    elif step[0][0] == 'D':
        print(step)
        y_dist = int(step[1:])
        print(y_dist)
        if len(combs) != 0:
            x_last = combs[len(combs)-1][0]
            y_last = combs[len(combs)-1][1]
        for y in range(1, y_dist+1):
            combs.append((x_last, y_last-y))
        print(len(combs))
        


    
#print (combs)
```
