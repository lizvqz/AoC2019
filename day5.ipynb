```python
filename = '/Users/liz/Documents/Advent2019/day5.txt'
v = []

with open(filename) as file:
    for line in file:
        for column in line.strip().split(','):
            v.append(int(column))                   

#print(v)

ptr = 0
in_3 = 5

    
def Value(ptr, a):
    if a == '1':
        return v[ptr]
    else:
        return v[v[ptr]]
    
def Result(ptr, a, result):
    if a == '1':
        v[ptr] = result
    else:
        v[v[ptr]] = result
        
def Output(ptr, a):
    if a == '1':
        print(v[ptr])
    else:
        print(v[v[ptr]])    

while True:
    
    print('opcode address: '+ str(ptr))
#    print(v)
    a = str(v[ptr]).zfill(5) # Example: '11002'

    if a[3:5] == '01':       
        result = Value(ptr+1, a[2:3]) + Value(ptr+2, a[1:2])
        Result(ptr+3, a[0:1], result)
        ptr+=4
    elif a[3:5] == '02':
        result = Value(ptr+1, a[2:3]) * Value(ptr+2, a[1:2])
        Result(ptr+3, a[0:1], result)
        ptr+=4
    elif a[3:5] == '99':
        break
    elif a[3:5] == '03':
        Result(ptr+1, a[2:3], in_3)
        ptr+=2
    elif a[3:5] == '04':
        Output(ptr+1, a[2:3])
        ptr+=2
    elif a[3:5] == '05':
        if Value(ptr+1, a[2:3]) != 0:
            ptr = Value(ptr+2, a[1:2])
        else:
            ptr+=3
    elif a[3:5] == '06':
        if Value(ptr+1, a[2:3]) == 0:
            ptr = Value(ptr+2, a[1:2])
        else:
            ptr+=3
    elif a[3:5] == '07':
        if Value(ptr+1, a[2:3]) < Value(ptr+2, a[1:2]):
            Result(ptr+3, a[0:1], 1)
        else:
            Result(ptr+3, a[0:1], 0)
        ptr+=4
    elif a[3:5] == '08':
        if Value(ptr+1, a[2:3]) == Value(ptr+2, a[1:2]):
            Result(ptr+3, a[0:1], 1)
        else:
            Result(ptr+3, a[0:1], 0)
        ptr+=4
        
    else:
        print('invalid opcode')
#        print(v)
        break

            
```
